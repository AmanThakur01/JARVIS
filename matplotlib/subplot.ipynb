```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
plt.subplot(2,2,1)
style.use('ggplot')
plt.scatter([9, 8, 7], [6, 5, 4], label='first', linewidth=4)
# plt.plot([1, 2, 3], [4, 5, 6], label='second', linewidth=2,color='b',marker='^',linestyle="--",markersize=10)
plt.plot([1, 2, 3], [4, 5, 6],"go--",markersize=10, linewidth=2, label='second')
plt.axis([0,10,2,7])
plt.title('random data',fontsize=15)
plt.ylabel('y-axis',fontsize=10)
plt.xlabel('x-axis',fontsize=10)
plt.legend(["second"],loc=4)
plt.grid(color='w',linestyle='-',linewidth=15)

plt.subplot(2,2,2)
class_name=['py','ML','AI','DS']
stds1=[12,10,6,7]
stds2=[17,16,36,27]
index=np.arange(len(class_name))
width=0.3
plt.barh(index,stds1,width,edgecolor='m')#,alpha=0.5
plt.barh(index+width,stds2,width,edgecolor='m')
plt.yticks(index+width,class_name,rotation=32)

plt.subplot(2,2,3)
plt.pie([2,2])

plt.subplot(2,2,4,projection='polar',facecolor='k',frameon=False)
plt.show()
```
