```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style
import random
```

```python
mlStd=np.random.randint(18,45,100)
pyStd=np.random.randint(8,30,100)
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.hist([mlStd,pyStd],rwidth=0.8,histtype="bar",label=["ML Student","Python Student"])#bins=default if we set bins called bar xvalue e.i. 10-20 bar x-axis
####hist(oriantation=vertical,align=middel,bins,color='r')
plt.title("Python and ML Students Number")
plt.xlabel("Age of Student")
plt.ylabel("No. of Student")
plt.legend()
plt.show()
```
