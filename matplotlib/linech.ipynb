```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
plt.scatter([9, 8, 7], [6, 5, 4], label='first', linewidth=4)
# plt.plot([1, 2, 3], [4, 5, 6], label='second', linewidth=2,color='b',marker='^',linestyle="--",markersize=10)
plt.plot([1, 2, 3], [4, 5, 6],"go--",markersize=10, linewidth=2, label='second')
plt.axis([0,10,2,7])
plt.title('random data',fontsize=15)
plt.ylabel('y-axis',fontsize=10)
plt.xlabel('x-axis',fontsize=10)
plt.legend(["second"],loc=4)
plt.grid(color='w',linestyle='-',linewidth=15)
plt.show()
```
